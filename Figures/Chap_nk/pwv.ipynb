```python
%matplotlib notebook
import matplotlib as mpl
mpl.rcParams['savefig.pad_inches'] = 0
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import readsav
import fkplotlib

fkplotlib.use_txfonts()

p = readsav("atm_pardo_2017.save", python_dict=True)
```

```python
fig, ax = plt.subplots(figsize=(5, 4))
pwvs_i_want = [0.5, 1.0, 2.0, 3.0, 4.0, 6.0]
for i, pwv in enumerate(pwvs_i_want):
    j = np.where(p["pwv"] == pwv)[0]
    tau = p["zopa"][:, j]
    tr = np.exp(-tau)
    label = f"{pwv:.0f} mm" if pwv >= 1.0 else f"{pwv:.1f} mm"
    ax.plot(p["freqs"], tr, color=plt.cm.magma(i / len(pwvs_i_want)), label=label)
fkplotlib.ax_bothticks(ax)
fkplotlib.ax_legend(
    ax, ncol=3, title="Vapeur d'eau prÃ©cipitable", loc=9, 
    handlelength=1, borderaxespad=0, columnspacing=.75, bbox_to_anchor=(0.5, 1.05))
ax.set_xlim(p["freqs"][0], 370)
ax.set_ylim(0.0, 1.2)
ax.set_xlabel(r"FrÃ©quence $\nu \;[{\rm GHz}]$")
ax.set_ylabel(r"Transmission atmosphÃ©rique au zÃ©nith")
fig.subplots_adjust(left=0.1, right=0.98, bottom=0.088, top=0.95)
fig.savefig("atm_trans.pdf")
```
